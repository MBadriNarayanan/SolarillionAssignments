## Task 1 - Plotting
* Plot the values of occupancy percentage versus number of weeks since release, for all movies in the dataset. 
* Plot the number of other releases in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.
* Plot the number of shows in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle as pkl

dataframe = pd.read_csv(r'C:\Users\badri\Desktop\MLBasics-master\Data\CinemaData.csv')

dataframe = dataframe.iloc[:,1:13]
 
dataframe 
```

```python
Y = dataframe['OccPer']
X = dataframe['WeeksSinceRelease']

plt.ylabel(" Occupancy Percentage ")
plt.xlabel(" Number Of Weeks Since Release ")
plt.scatter(X,Y,color = 'red')
plt.show()

Y = dataframe['OtherReleasesInWeek']
X = dataframe['OccPer']
plt.ylabel(" Number Of Other Releases In The Week ")
plt.xlabel(" Occupancy Percentage ")
plt.scatter(X,Y,color = 'blue')
plt.show()

Y = dataframe['OtherReleasesInWeek']
X = dataframe['Lifetime'] - dataframe['WeeksSinceRelease']
plt.ylabel(" Number Of Other Releases In The Week ")
plt.xlabel(" Weeks Until The End Of Lifetime ")
plt.scatter(X,Y,color = 'green')
plt.show()

Y =  dataframe['ShowsInWeek']
X = dataframe['OccPer']
plt.ylabel(" Number Of Shows In The Week ")
plt.xlabel(" Occupancy Percentage ")
plt.scatter(X,Y,color = 'black')
plt.show()

Y = dataframe['ShowsInWeek']
X = dataframe['Lifetime'] - dataframe['WeeksSinceRelease']
plt.ylabel(" Number Of Shows In The Week ")
plt.xlabel(" Weeks Untill The End Of Lifetime ")
plt.scatter(X,Y,color = 'brown')
plt.show()
```
